# Question 3: Pitcher A vs Pitcher B Simulation

```python
import random
import pandas as pd
import numpy as np

def pitcher_50_50():
    # Generate a random number (0 or 1) representing heads or tails
    result = random.random()

    # Return 'Heads' for 0, 'Tails' for 1
    return 'K' if result >= 0.5 else 'BB'

def bip_outcome():
    home_run = random.random()
    if(home_run<0.036):
        return 'HR'
    
    random_number = random.random()
    
    if(random_number<0.3):
        random_hit = random.random()
        if(random_hit<0.25):
            return "2B"
        else:
            return "1B"
    else:
        return "Out"
    

# Simulate a coin flip
result = pitcher_50_50()
print(f"The result of the coin flip is: {result}")
```

```python
runs_allowed = []
iterations = 1000000

for i in range(iterations):
    inning_over = 0
    num_k = 0
    num_bb = 0
    
    while(not inning_over):
        result = pitcher_50_50()
        if(result == 'K'):
            num_k+=1
        else:
            num_bb+=1
            
        if(num_k==3):
            inning_over = 1
            
    runs_allowed.append(num_bb//4)
```

```python
np.mean(runs_allowed)
```

```python
iterations = 1000000

def run_simulation(iterations, show_output):
    for i in range(iterations):
        runs_allowed_2 = []

        if(show_output):
            print("Inning: " + str(i))
        inning_over = 0
        num_out = 0
        run_on = 0
        bases = 0

        runs_scored = 0

        while(not inning_over):
            result = bip_outcome()        
            if(result == 'Out'):
                num_out+=1

            if(result == 'HR'):
                runs_scored += (1+run_on)
                run_on = 0
                bases = 0

            if(result == "1B"):
                run_on += 1
                bases += 1

                scored = bases//4

                bases = bases%4
                runs_scored += scored
                run_on -= scored

            if(result == "2B"):
                run_on += 1
                bases += 2

                scored = bases//4

                bases = bases%4
                runs_scored += scored
                run_on -= scored

            if(result == "3B"):
                run_on += 1
                bases += 3

                scored = bases//4

                bases = bases%4
                runs_scored += scored
                run_on -= scored

            if(num_out==3):
                inning_over = 1

            if(show_output):
                print("Result: " + result)
                print("Runs Scored: " + str(runs_scored))
                print("Runners On:" + str(run_on))
                print('\n')

        runs_allowed_2.append(runs_scored)
        if(show_output):
            print('\n')
            print('\n')    
    return runs_allowed_2
```

```python
runs_allowed_sim = run_simulation(1000000,0)
np.mean(runs_allowed_2)
```
